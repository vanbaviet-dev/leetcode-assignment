```python
import pandas as pd 
```

```python
data = [[1, "a@b.com"], [2, "c@d.com"], [3, "a@b.com"]]
person = pd.DataFrame(data, columns=["id", "email"]).astype(
    {"id": "Int64", "email": "object"}
)
```

```python
def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    df = person.groupby("email").size().reset_index(name="cnt")
    return df.loc[df["cnt"] > 1][["email"]].rename(columns={"email": "Email"})
```

```python
print(duplicate_emails(person=person))
```
