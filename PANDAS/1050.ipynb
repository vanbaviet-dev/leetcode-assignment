```python
import pandas as pd
```

```python
data = [[1, 1, 0], [1, 1, 1], [1, 1, 2], [1, 2, 3], [1, 2, 4], [2, 1, 5], [2, 1, 6]]
actor_director = pd.DataFrame(
    data, columns=["actor_id", "director_id", "timestamp"]
).astype({"actor_id": "int64", "director_id": "int64", "timestamp": "int64"})
```

```python
def actors_and_directors(actor_director: pd.DataFrame) -> pd.DataFrame:
    actor_director["rank"] = actor_director.groupby(["actor_id", "director_id"])[
        "timestamp"
    ].rank(method="first", ascending=True)
    return actor_director[actor_director["rank"] == 3][["actor_id", "director_id"]]
```

```python
print(actors_and_directors(actor_director=actor_director))
```
