```python
import pandas as pd
```

```python
data = [[1, '2017-01-01', 10], [2, '2017-01-02', 109], [3, '2017-01-03', 150], [4, '2017-01-04', 99], [5, '2017-01-05', 145], [6, '2017-01-06', 1455], [7, '2017-01-07', 199], [8, '2017-01-09', 188]]
stadium = pd.DataFrame(data, columns=['id', 'visit_date', 'people']).astype({'id':'Int64', 'visit_date':'datetime64[ns]', 'people':'Int64'})
```

```python
def human_traffic(stadium: pd.DataFrame) -> pd.DataFrame:
    stadium = stadium[stadium["people"] >= 100]
    stadium = stadium.sort_values(by="id", ascending=True)
    stadium["prev_id_2"] = stadium["id"].shift(2)
    stadium["prev_id_1"] = stadium["id"].shift(1)
    stadium["next_id_1"] = stadium["id"].shift(-1)
    stadium["next_id_2"] = stadium["id"].shift(-2)
    mask_1 = (stadium["id"] - stadium["prev_id_2"] == 2) & (
        stadium["id"] - stadium["prev_id_1"] == 1
    )
    mask_2 = (stadium["id"] - stadium["prev_id_1"] == 1) & (
        stadium["next_id_1"] - stadium["id"] == 1
    )

    mask_3 = (stadium["next_id_1"] - stadium["id"] == 1) & (
        stadium["next_id_2"] - stadium["id"] == 2
    )
    mask = mask_1 | mask_2 | mask_3
    return (
        stadium[mask][["id", "visit_date", "people"]]
        .sort_values(by="visit_date", ascending=True)
        .reset_index(drop=True)
    )
```

```python
print(human_traffic(stadium))
```
