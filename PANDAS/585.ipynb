```python
import pandas as pd
```

```python
data = [[1, 10, 5, 10, 10], [2, 20, 20, 20, 20], [3, 10, 30, 20, 20], [4, 10, 40, 40, 40]]
insurance = pd.DataFrame(data, columns=['pid', 'tiv_2015', 'tiv_2016', 'lat', 'lon']).astype({'pid':'Int64', 'tiv_2015':'Float64', 'tiv_2016':'Float64', 'lat':'Float64', 'lon':'Float64'})
```

```python
def find_investments(insurance: pd.DataFrame) -> pd.DataFrame:
    # #Solution 1
    # tiv_2015 = insurance.groupby("tiv_2015").filter(lambda x: len(x) > 1)[["pid"]]
    # tiv_2016 = insurance.groupby(["lat", "lon"]).filter(lambda x: len(x) == 1)[
    #     ["pid", "tiv_2016"]
    # ]
    # total = pd.merge(tiv_2015, tiv_2016, on="pid", how="inner")["tiv_2016"].sum()
    # rounded_total = round(total, 2)
    # result = pd.DataFrame({"tiv_2016": [rounded_total]})
    # return result
    # Solution 2
    tiv_2015 = insurance.duplicated(subset=["tiv_2015"], keep=False)
    tiv_2016 = ~insurance.duplicated(subset=["lat", "lon"], keep=False)
    sum = insurance.loc[tiv_2015 & tiv_2016, "tiv_2016"].sum()
    result = pd.DataFrame({"tiv_2016": [round(sum, 2)]})
    return result
```

```python
print(find_investments(insurance))
```
