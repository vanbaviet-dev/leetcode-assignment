```python
import pandas as pd
```

```python
data = [[8], [8], [3], [3], [3]]
my_numbers = pd.DataFrame(data, columns=['num']).astype({'num':'Int64'})
```

```python
def biggest_single_number(my_numbers: pd.DataFrame) -> pd.DataFrame:
    m = my_numbers["num"].duplicated(keep=False)
    numbers_single = my_numbers[~m].sort_values(by="num", ascending=False).head(1)
    if numbers_single.empty:
        return pd.DataFrame([[None]], columns=["num"]).astype({"num": "Int64"})
    return numbers_single
```

```python
print (biggest_single_number(my_numbers))
```
