```python
import pandas as pd
```

```python
data = [[1, 'Joe', 70000, 1], [2, 'Jim', 90000, 1], [3, 'Henry', 80000, 2], [4, 'Sam', 60000, 2], [5, 'Max', 90000, 1]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'departmentId':'Int64'})
data = [[1, 'IT'], [2, 'Sales']]
department = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
```

```python

def department_highest_salary(
    employee: pd.DataFrame, department: pd.DataFrame
) -> pd.DataFrame:
    mask = employee["salary"].eq(
        employee.groupby("departmentId")["salary"].transform("max")
    )
    df = employee[mask].merge(
        department, left_on="departmentId", right_on="id", suffixes=("_emp", "_dp")
    )
    return df.rename(
        columns={"name_emp": "Employee", "name_dp": "Department", "salary": "Salary"}
    )[["Department", "Employee", "Salary"]]
```

```python
print(department_highest_salary(employee=employee,department=department))
```
