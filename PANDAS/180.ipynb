```python
import pandas as pd
```

```python
data = [[1, 1], [2, 1], [3, 1], [4, 2], [5, 2], [6, 2], [7, 2]]
logs = pd.DataFrame(data, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})
```

```python
def consecutive_numbers(logs: pd.DataFrame) -> pd.DataFrame:
    logs["prev_ num"] = logs["num"].shift(1)
    logs["nxt_num"] = logs["num"].shift(-1)
    return (
        logs.loc[
            (logs["num"] == logs["prev_ num"]) & (logs["num"] == logs["nxt_num"]),
            ["num"],
        ]
        .drop_duplicates()
        .rename(columns={"num": "ConsecutiveNums"})
    )
```

```python
print(consecutive_numbers(logs))
```
