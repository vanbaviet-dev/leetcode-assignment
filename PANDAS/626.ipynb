```python
import pandas as pd
```

```python
data = [[1, 'Abbot'], [2, 'Doris'], [3, 'Emerson'], [4, 'Green'], [5, 'Jeames'],[6, 'JAeames']]
seat = pd.DataFrame(data, columns=['id', 'student']).astype({'id':'Int64', 'student':'object'})
```

```python
def exchange_seats(seat: pd.DataFrame) -> pd.DataFrame:
    seat['nxt_id'] = seat['id'].shift(-1)
    seat['prev_id'] = seat['id'].shift(1)
    seat['id'] = seat.apply(lambda row: (row['id'] if pd.isna(row['nxt_id'])==True else row['nxt_id']) if row['id'] % 2 != 0 else row['prev_id'], axis=1)
    return seat[['id', 'student']].sort_values('id').reset_index(drop=True)
```

```python
print(exchange_seats(seat))
```
