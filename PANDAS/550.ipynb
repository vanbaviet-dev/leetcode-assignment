```python
import pandas as pd
```

```python
data = [
    [1, 2, "2016-03-01", 5],
    [2, 3, "2017-06-25", 1],
    [3, 1, "2016-03-02", 0],
    [3, 4, "2018-07-03", 5],
    [1, 2, "2016-03-02", 6],
]
activity = pd.DataFrame(
    data, columns=["player_id", "device_id", "event_date", "games_played"]
).astype(
    {
        "player_id": "Int64",
        "device_id": "Int64",
        "event_date": "datetime64[ns]",
        "games_played": "Int64",
    }
)
```

```python
def gameplay_analysis(activity: pd.DataFrame) -> pd.DataFrame:
    activity = activity.sort_values(by=["player_id", "event_date"])
    activity["nxt_date"] = activity.groupby("player_id")["event_date"].shift(-1)
    activity["first_date"] = activity.groupby("player_id")["event_date"].transform(
        "min"
    )
    fraction = len(
        activity.loc[
            (activity["event_date"] == activity["first_date"])
            & (activity["event_date"] + pd.Timedelta(days=1) == activity["nxt_date"])
        ]
    )
    users = len(activity["player_id"].drop_duplicates())
    return pd.DataFrame(
        {"fraction": pd.Series([round(fraction / users, 2)], dtype="float")}
    )
```

```python
print(gameplay_analysis(activity=activity))
```
