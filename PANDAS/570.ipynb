```python
import pandas as pd
```

```python
data = [
    [101, "John", "A", None],
    [102, "Dan", "A", 101],
    [103, "James", "A", 101],
    [104, "Amy", "A", 101],
    [105, "Anne", "A", 101],
    [106, "Ron", "B", 101],
]
employee = pd.DataFrame(data, columns=["id", "name", "department", "managerId"]).astype(
    {"id": "Int64", "name": "object", "department": "object", "managerId": "Int64"}
)
```

```python
def find_managers(employee: pd.DataFrame) -> pd.DataFrame:
    values = employee.groupby("managerId").count().reset_index()
    values = values[values["id"] > 4]
    result = employee[employee["id"].isin(values["managerId"])]
    return result[["name"]]
```

```python
print(find_managers(employee=employee))
```
