```python
import pandas as pd
```

```python
data = [[1, 2, '2016/06/03'], [1, 3, '2016/06/08'], [2, 3, '2016/06/08'], [3, 4, '2016/06/09']]
request_accepted = pd.DataFrame(data, columns=['requester_id', 'accepter_id', 'accept_date']).astype({'requester_id':'Int64', 'accepter_id':'Int64', 'accept_date':'datetime64[ns]'})
```

```python
def most_friends(request_accepted: pd.DataFrame) -> pd.DataFrame:
    df1 = request_accepted.copy()
    df2 = request_accepted.copy()

    df1 = df1.groupby("accepter_id").size().reset_index(name="num")
    df2 = df2.groupby("requester_id").size().reset_index(name="num")
    df = pd.concat(
        [
            df1.rename(columns={"accepter_id": "id"}),
            df2.rename(columns={"requester_id": "id"}),
        ]
    )
    df = (
        df.groupby("id")["num"]
        .sum()
        .reset_index()
        .sort_values(by="num", ascending=False)
        .head(1)
    )
    return df
```

```python
print(most_friends(request_accepted))
```
