```python
import pandas as pd
```

```python
data = [[1, 3.5], [2, 3.65], [3, 4.0], [4, 3.85], [5, 4.0], [6, 3.65]]
scores = pd.DataFrame(data, columns=["id", "score"]).astype(
    {"id": "Int64", "score": "Float64"}
)
```

```python
def order_scores(scores: pd.DataFrame) -> pd.DataFrame:
    result = scores.sort_values("score", ascending=False)
    result["rank"] = result["score"].rank(method="dense", ascending=False)
    return result[["score", "rank"]]
```

```python
print(order_scores(scores=scores))
```
