```python
import pandas as pd
```

```python
data = [[1, 5], [2, 6], [3, 5], [3, 6], [1, 6]]
customer = pd.DataFrame(data, columns=['customer_id', 'product_key']).astype({'customer_id':'Int64', 'product_key':'Int64'})
data = [[5], [6]]
product = pd.DataFrame(data, columns=['product_key']).astype({'product_key':'Int64'})
```

```python
def find_customers(customer: pd.DataFrame, product: pd.DataFrame) -> pd.DataFrame:
    cnt_product = product["product_key"].nunique()
    remove_duplicates = customer.drop_duplicates(subset=["customer_id", "product_key"])
    cnt_customer = (
        remove_duplicates.groupby("customer_id").size().reset_index(name="cnt")
    )
    result = cnt_customer[cnt_customer["cnt"] == cnt_product][["customer_id"]]
    return result
```

```python
print(find_customers(customer, product))
```
