```python
import pandas as pd
```

```python
data = [[1, None], [2, 1], [3, 1], [4, 2], [5, 2]]
tree = pd.DataFrame(data, columns=['id', 'p_id']).astype({'id':'Int64', 'p_id':'Int64'})
```

```python
def tree_node(tree: pd.DataFrame) -> pd.DataFrame:
    tree["type"] = tree.apply(
        lambda row: (
            "Root"
            if pd.isna(row["p_id"])
            else ("Inner" if row["id"] in tree["p_id"].values else "Leaf")
        ),
        axis=1,
    )
    return tree[["id", "type"]]
```

```python
print(tree_node(tree))
```
